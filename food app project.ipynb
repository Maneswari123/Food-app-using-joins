# points to remember before joining the data sets.
#### 1) To join two data sets the primary requisite is primary key.
#### 2) Atleast one column values should match in both the data sets.
#### 3) we can not merge two trasactional data sets.
#### 4) The primary key should be unique.
#### 5) Atleast one data set should not have duplicate records.

## joining the data sets.

```python
data1=pd.merge(cust,orders,on="customer_id",how="left")
data1
```

```python
data1.dtypes
```

```python
data1.nunique()
```

```python
data2=pd.merge(data1,drivers,on="driver_id",how="left")
data2
```

```python
data2.dtypes
```

```python
data3=pd.merge(data2,order_items,on="order_id",how="left")
data3
```

```python
data3.dtypes
```

```python
data4=pd.merge(data3,food_items,on="item_id",how="left")
data4
```

```python
final_aggregated_data=pd.merge(data4,rest,on="restaurant_id",how="left")
final_aggregated_data
```
